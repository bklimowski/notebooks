**Overview** In this article You can find description and implementation of technique from paper *From time series to complex networks: The visibility graph*[^1]. Graph representation unables extraction of useful properties like irreversibility of time series [^2].  
 


##  Introduction Pozdro dla Gosi

On the beggining we can ask question: Why is it useful to shift one data representation to another?  

Foremost, it enables us to apply analysis techniques unique for recived entity. As a example, after translating time series into image, we can use well developed and succesful deep learning methods in our analysis [^3].


![img](TimeSeriesToImage.png)
**Figure 1** *Translating time series into image*[^3]


Additionaly, right method can get rid of insignificant information, preserving only useful properties. For example, technique called *symbolic dynamics*, translate time series into set of characters. Analysis of recived sequences can give us unique information about time series dynamics. [^4]

Translating series into visibility graph can be useful for extraction of time series features, which characterisation is really hard with *classical* methods. Additionaly, this approach share many similarities with symbolic dynamics [^1], it also drop many redundant properites of series, focusing only on important part of it.



## **Visibility graph**
Idea behind this approach is very simple. Lets take time series X(t), every point of it is become node of our graph. Directed connections between them are established on basis of *visibility*. How we define it? Lest draw vertical line to every point of series, if we can connect line of point to another line before him, without crossing another, they are connected. It is crucial that we connect lines only in forward direction. After checking *visibility* of every point we have two features characterizing node:  
<center>$k_{in}$ - how many other nodes *sees* it  </center>
<center>$k_{out}$ - how many other nodes it *sees*</center>


![img](SeriesToGraph.png)

**Figure 2** *Series to graph* [^2]  

On the basis of these two properties we can create a marker of irreversibility. Asymmetric function should have diffirend distributions of $k_{in}$ and $k_{out}$,we can quantify this dissimilarity with Kullback–Leibler divergence:

$$ D[P_{out}(k)|| P_{in}(k)] = \sum_{k} P_{out} \log \frac{P_{out}(k)}{P_{in}(k)}$$

Lest's see how it's works in pratice.

## **Creating visibility graph**

Below is implementation of visibility graph in form of class, only thing it needs to work is time series. Detailed description of every variable is in code.   
Only thing worth explaining is approach used in method checking visibility (create_connections_dict):
1. Get point N of time series.
2. In front of our point find first point greter than N. If N is maximum in this interval take point next to him in value. Recived point is end of our area of interest.
3. Create empty list which will contains not visible points, add points lesser than point next to N.
4. From derivative of region of interest, select points with value greater than zero and going forward add to list point lesser than them.
5. Connect with every point of region of interest, excluded those from list.


```python
%pylab inline
pylab.rcParams['figure.figsize'] = (12, 8)
import numpy as np
import pandas as pd
import networkx as nx
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import signal
plt.style.use('fivethirtyeight')
```

```python
class VisibilityGraph(object):
    
    '''
    This class implement visibility graph representation of time series. 
    Only thing needed in input is time series itself, after initilaization code creates class variables:
    ts - time series
    edges_dict - visibility of time series points
    graph_represantation - networkX graph class with visibility graph
    k_dist_in - distribution of k_in
    k_dit_out - distribution of k_out
    KL_divergence - Kullback–Leibler divergence calculated with k_dist_in and k_dist_out
    
    '''
    
    
    def __init__(self,time_series):
        self.ts = pd.Series(time_series)
        self.edges_dict = self.create_connections_dict()
        self.graph_representation = nx.DiGraph(self.edges_dict)
        self.k_dict = self.k_values()
        self.k_dist_in = pd.value_counts(self.k_dict['k_in']).sort_index()/np.sum(self.k_dict['k_in'])
        self.k_dist_out = pd.value_counts(self.k_dict['k_out']).sort_index()/np.sum(self.k_dict['k_out'])
        self.KL_divergence = np.sum(self.k_dist_in*(np.log(self.k_dist_in/self.k_dist_out)))
        
    def create_connections_dict(self):
        edges_dict = {}
        ts_diff = self.ts.diff()

        for ii in range(self.ts.shape[0]-1):

            if self.ts[ii:].max() == self.ts[ii]:
                end = self.ts[ii+1:].sort_values(ascending=False).index[0]
            else:
                end = self.ts[ii:][self.ts>self.ts[ii]].index[0]

            tmp_diff = ts_diff[ii+1:end]
            tmp_diff[tmp_diff>0]
            to_remove = []
            to_remove.extend(self.ts[ii+1:end][self.ts[ii+1:end]<self.ts[ii+1]].index)

            for jj in tmp_diff[tmp_diff>0].index:
                to_remove.extend(self.ts[jj:end][self.ts[jj:end]<self.ts[jj]].index)

            to_remove = list(set(to_remove))
            edges_list = [x for x in self.ts[ii+1:end].index if x not in to_remove]
            edges_list.append(end)
            edges_dict[ii] = edges_list

        return edges_dict
    
    def k_values(self):
        k_in_values, k_in, k_out = [],[],[]
        for ii in self.edges_dict.keys():
            k_out.append(len(self.edges_dict[ii]))
            k_in_values.extend(self.edges_dict[ii])

        for jj in self.edges_dict.keys():
            k_in.append(k_in_values.count(jj))
            
        k_dict = {'k_in':k_in[1:],'k_out':k_out}
        
        return k_dict
```

## Simple example - sin weave
Let's see how it works on basic weaveform.

```python
x = np.linspace(0,4*np.pi,100)
sin_ts =  np.cos(x*2)
```

```python
plt.plot(x,sin_ts)
plt.xlabel('x')
plt.ylabel('sin(x)')
```

Initializign class

```python
ts_visibility_graph = VisibilityGraph(sin_ts)
```

In fact evrything is calculated now, with networkX package we can draw created graph.

```python
nx.draw_circular(ts_visibility_graph.graph_representation)
```

As we can see, recived graph is quite regular, even symetrical.

```python
width = np.min(np.diff(ts_visibility_graph.k_dist_out.keys()))/3 
plt.bar(ts_visibility_graph.k_dist_in.keys(),ts_visibility_graph.k_dist_in.values,width)
plt.bar(ts_visibility_graph.k_dist_out.keys() + width,ts_visibility_graph.k_dist_out.values,width)
plt.xlabel('Number of connections')
plt.ylabel('Density')
```

```python
plt.plot(np.log2(ts_visibility_graph.k_dist_in),'-.b')
plt.plot(np.log2(ts_visibility_graph.k_dist_out),'-.r')
plt.xlabel('Number of connections')
plt.ylabel('Log2 of density')
print('KL divergence:',str(np.round(ts_visibility_graph.KL_divergence,3)))
```

### Kullback–Leibler divergence of connections distributions as a mesure of time series irreversibility
One of nonlinear property of time series is irreversibility. In layman term, we can connect irreversibility with symmetry, when mirror reflection of time series is different from original, series is irreversible. But this simple definiton doesnt simply apply in practicie, in realilty it's really hard to quantify this property

### Nonliniear maps

Lets create some time series for testing. For our analisys it's siutable to take nonlinear time series like logistic, circle and gauss maps, irreversible stationary series as saw signal and simple sin with gaussian noise.

```python
def logistic_map(n,r,a):
    x_n = a
    log_ts = []
    for ii in range(n):
        x_n = r*x_n*(1-x_n)
        log_ts.append(x_n)
    return pd.Series(log_ts)

def cirlce_map(n,omega,K,a):
    x_n = a
    log_ts = []
    for ii in range(n):
        x_n = x_n + omega - np.mod(K*np.sin(np.pi*x_n)/np.pi,1)
        log_ts.append(x_n)
    return pd.Series(log_ts)

def gauss_map(n,alpha,beta,a):
    x_n = a
    log_ts = []
    for ii in range(n):
        x_n = np.mod(np.exp(-alpha*x_n**2),1)+beta
        log_ts.append(x_n)
    return pd.Series(log_ts)
```

```python
circle_map_ts = cirlce_map(100,0.6633,0.84,0.1)
gauss_map_ts = gauss_map(1000,6.111,-0.4824,0.55)
logistic_map_ts = logistic_map(5000,3.7122233,0.23)

time= pd.Series(np.arange(1000))
random_ts = pd.Series(np.random.rand(1000))
saw_ts = pd.Series(signal.sawtooth(time)) + random_ts/4
sin_with_noise_ts = np.sin(time) + random_ts/4

ts_list = [circle_map_ts,gauss_map_ts,logistic_map_ts,
           random_ts,saw_ts,sin_with_noise_ts]

ts_list_names = ['circle_map_ts','gauss_map_ts','logistic_map_ts',
                 'random_ts','saw_ts','sin_with_noise_ts']

for ii in range(0,len(ts_list)):
    
    plt.figure(1,figsize=(20,20))
    
    plt.subplot(6,2,2*(ii+1))
    plt.plot(ts_list[ii][:50].tolist()[::-1])
    plt.title(('reversed '+ts_list_names[ii]))

    plt.subplot(6,2,(2*(ii+1)-1))
    plt.plot(ts_list[ii][:50])
    plt.title(ts_list_names[ii])
```

```python
for ii in range(len(ts_list)):
    visibility_graph = VisibilityGraph(ts_list[ii])
    
    
    plt.figure(1,figsize=(15,10))
    plt.subplot(3,2,ii+1)
    plt.plot(np.log2(visibility_graph.k_dist_out),'-.b')
    plt.plot(np.log2(visibility_graph.k_dist_in),'-.r')
    title_str = str(ts_list_names[ii]) + ': KL_div = ' + str(np.round(visibility_graph.KL_divergence,2))
    plt.title(title_str)
```
