```python
series = np.sin(np.arange(40)) + np.random.rand(40) + np.cos(np.arange(40)/2) + np.random.rand(40)*2
series = (series + np.abs(series.min()))
series = series/series.max()
plt.plot(series)
```

```python
train = pd.DataFrame()
for ii in range(series.shape[0]-5):
    train[ii] = series[ii:ii+6]
train = train.transpose()
```

```python
train_x  =train.iloc[:25,:-1].values
train_y  =train.iloc[:25,-1].values

test_x  =train.iloc[25:,:-1].values
test_y  =train.iloc[25:,-1].values
```

```python
model = Sequential()
# Dense(64) is a fully-connected layer with 64 hidden units.
# in the first layer, you must specify the expected input data shape:
# here, 20-dimensional vectors.
model.add(Dense(64, activation='relu', input_dim=5))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, kernel_initializer='normal'))

model.compile(loss='mean_squared_error', optimizer='adam')

model.fit(train_x, train_y,
          epochs=150,
          batch_size=128)
score = model.evaluate(test_x, test_y, batch_size=128)
```
